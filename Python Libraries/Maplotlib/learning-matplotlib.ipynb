# **Bar Graph**

```python
x = np.arange(1,6)
cars1 = [23, 14, 8, 21, 10]
cars2 = [13, 25, 28, 6, 10]
```

```python
plt.bar(x, cars1, width = 0.3, color = 'red', tick_label = [2017, 2018, 2019, 2020, 2021]) 
# width - changes the width of the bar
# tick_label - adds label names to the x axis 
plt.bar(x, cars2, width = 0.3)
plt.show()
```

```python
plt.bar(x, cars1, width = 0.3, color = 'red', tick_label = [2017, 2018, 2019, 2020, 2021],label = 'Car 2') 
plt.bar(x + 0.4, cars2, width = 0.3 , label = 'Car 2') 
# x + 0.3 shifts the second graph by 0.4 to the right, thus displaying the graphs side by side
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales")
plt.xlim(0, 6)
plt.show()
```

# **Pie Chart**

```python
subjects = ["Computer", "English", "Maths", "Chemistry"]
marks = np.array([78, 45, 60, 90])
```

```python
plt.pie(x = marks, labels = subjects)
plt.show()
```

```python
# formatting 
plt.pie(x = marks, labels = subjects, explode = [0, 0, 0.3, 0], shadow = True, autopct = "%0.2f")
# explode - to separate particular datas from the pie graph
# shadow - adds shadow to the entire graph
# autopct - provides the percentage of each data in the graph
plt.show()
```

# **Histogram**

```python
# generating values for a histogram 
x_ = np.random.randn(500)
maths_mu = 70
maths_snd = 8
maths = np.round(x_ * maths_snd + maths_mu)

phys_mu = 50
phys_snd = 10
phys = np.round(x_ * phys_snd + phys_mu)
```

```python
print(maths.max())
print(maths.min())
print(phys.max())
print(phys.min())
```

```python
plt.hist(maths, bins = 10)
# bins: divide the total value set into 10 subsets
# each bar graph is the total frequency of all the values in that particular subsets
plt.show()
```

```python
plt.hist(maths, bins = 10)
plt.hist(phys, bins = 10, alpha = 0.7)
plt.show()
```

# **Subplot**

```python
x = np.arange(0, 10, 0.1)
y1 = x ** 2
y2 = x ** 3
y3 = np.sin(x)
y4 = np.cos(x)
```

```python
plt.subplot(2,2,1)
plt.plot(y1)
# 2,2 indicates it will take up 2 * 2 cells for the graph
# 1 means that if the entire plot is divided into 4 subplots, this graph takes up the first plot

plt.subplot(2,2,2)
plt.plot(y2)

plt.subplot(2,2,3)
plt.plot(y3)

plt.subplot(2,2,4)
plt.plot(y4)

plt.show()
```

# **3D Plot**

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
a = np.arange(4)
b = np.arange(4,7)
```

```python
a, b = np.meshgrid(a,b)
```

```python
a
```

```python
b
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_surface(a, b, a+b) # x-axis, y-axis, z-axis
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot_surface(a, b, np.array([[4]])) # x-axis, y-axis, z-axis
plt.show()
```
